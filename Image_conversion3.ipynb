```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from PIL import Image
```

```python
img = cv2.imread("/content/plant.jpg");
img=cv2.resize(img,(300,300))
cv2_imshow(img)
```

```python
#Conversion of RGB To GRAY
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
cv2_imshow(img_gray)
```

```python
#conversion of BGR to GRAY
bgrtogray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2_imshow(bgrtogray)
```

```python
#conversion of GRAY to BGR
img_BGR=cv2.cvtColor(bgrtogray,cv2.COLOR_GRAY2BGR)
cv2_imshow(img_BGR)
```

```python
#Conversion RGB to HSV MODEL
img_HSV=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
cv2_imshow(img_HSV)
```

```python
#Conversion HSV to RGB MODEL
img_RGB=cv2.cvtColor(img_HSV,cv2.COLOR_HSV2RGB)
cv2_imshow(img_RGB)
```

```python
#conversion BGR to  HSV MODEL
 
BGR_TO_HSV=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2_imshow(BGR_TO_HSV)
```

```python
#conversion HSV to  BGR MODEL
 
img_BGR=cv2.cvtColor(BGR_TO_HSV,cv2.COLOR_HSV2BGR)
cv2_imshow(img_BGR)
```

```python
#conversion RGB  to CMY model
R=100;
G=200;
B=245;
R1=(R/255)
R2=(G/255)
B1=(B/255)
C=1-R1
M=1-R2
Y=1-B 
color=(C,M,Y)
print(color)
```

```python
#removing the white color for image 
color=(255,255,255);
imgp=img-color
plt.imshow(imgp)
```
