```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/shubham.jpeg");
plt.imshow(img) 
```

```python
#Conversion of RGB To GRAY
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
```

```python
backtorgb = cv2.cvtColor(img_gray,cv2.COLOR_GRAY2RGB)
plt.imshow(backtorgb)
```

```python
#Conversion RGB to HSV MODEL
img_HSV=cv2.cvtColor(img,cv2.COLOR_RGB2HSV)
plt.imshow(img_HSV)
```

```python
#conversion BGR to  HLS MODEL
 
img_HSV=cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
plt.imshow(img_HSV)
```

```python
#conversion RGB to  HLS MODEL
 
img_HSV=cv2.cvtColor(img,cv2.COLOR_RGB2HLS)
plt.imshow(img_HSV)
```

```python
#conversion RGB  to CMY model
img_cmy=1-img
plt.imshow(img_cmy)
```

```python
#conversion CMY MODEL to RGB
img_RGB=1-img_cmy
plt.imshow(img_RGB)
```

```python
#conversion CMY to 
color=(137, 207, 240);
imgp=color-img
plt.imshow(imgp)
```

```python
#removing the white color for image 
color=(255,255,255);
imgp=img-color
plt.imshow(imgp)
```

```python
#removing the black for image
color=(1.9473, 1.9473, 1.9473)
imgp=img-color
plt.imshow(imgp)
```
